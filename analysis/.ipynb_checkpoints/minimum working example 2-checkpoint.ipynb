Now let's consider the conditional case. Similar to before, we consider 40 observations of 4-bid auctions, where each bid is drawn from a normal distribution of variance 1 and mean equal to 2 times an auction-specific covariate `x`, which ranges from 0 to 1 and is distributed uniformly. We also define the `get_cdf` function here again.

```python
from scipy import stats, special, integrate, interpolate
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
import statsmodels.nonparametric.kernel_density as kernel_density

def cdf(x, order_statistic_cdf, i, n):
        t = sp.Symbol('t')
        f = sp.Symbol('f')
        outside = np.math.factorial(n)/(np.math.factorial(i-1)*np.math.factorial(n-i))
        integrand = (t**(i-1)) * ((1-t)**(n-i))

        func = sp.integrate(integrand, (t,0,f))
        exp = outside*func-order_statistic_cdf(x)
        try:
            ans = [s for s in sp.real_roots(exp) if 0<=s<=1][0]
        except:
            ans = 1

        return float(ans)

def get_cdf(order_statistic_cdf, i, n):
    return lambda x: cdf(x, order_statistic_cdf, i, n)


class FastMean(stats.rv_continuous):
    
    def __init__(self, discrete_cdf, values):
        super().__init__()
        self.discrete_cdf = discrete_cdf
        
    def _cdf(self, x):
        cdf = interpolate.interp1d(values, self.discrete_cdf, bounds_error=False, fill_value=(0,1))
        return cdf(x)
    
    def _ppf(self, y):
        ppf = interpolate.interp1d(self.discrete_cdf, values, bounds_error=False, fill_value="extrapolate")
        return ppf(y)
```

We first generate the auction-specific covariates and the associated bids:

```python
covariates = np.random.uniform(0, 1, size=400)

bids = [stats.norm.rvs(loc=2*x, scale=1, size=4) for x, _ in zip(covariates, range(400))]
```

Again, we then consider the observed order statistics of the bids:

```python
one_four = [sorted(listing)[0] for listing in bids]
two_four = [sorted(listing)[1] for listing in bids]
three_four = [sorted(listing)[2] for listing in bids]
four_four = [sorted(listing)[3] for listing in bids]
```

Then we need to estimate the CDF of conditional distributions, that is, the distribution of bids given a particular realisation of `x`. This is therefore a 2-dimensional: we need to estimate separately the joint PDF of `bids` and `x`, and divide that by the marginal PDF of `x` in order to obtain the density of `bids` conditional on `x`. We first estimate the density of `x` using kernels, avoiding biases at the boundary using logit as an invertible mapping so that the support of the covariates is the entirety of $\mathbb{R}$. We visualise what this looks like:

```python
covariates_transformed = special.logit(covariates)
y_kde = stats.gaussian_kde(covariates_transformed)
x_kde = lambda x: y_kde.pdf(x)/(special.expit(x)*(1-special.expit(x)))

values = [x_kde(x) for x in np.linspace(0,1,num=40)]
plt.plot(np.linspace(0,1,num=40), values)
plt.ylim(0,2)
```

Next we estimate the joint density between bids and covariates, using a 2-dimensional Gaussian kernel. We again employ the `gaussian_kde` method. Again, for a given `x`, we estimate the marginal density of each of the four order statistics individually, and then combine them as before.

```python
data_pairs_one= [[covariates[idx], one_four[idx]] for idx, cov in enumerate(covariates)]
data_pairs_two= [[covariates[idx], two_four[idx]] for idx, cov in enumerate(covariates)]
data_pairs_three= [[covariates[idx], three_four[idx]] for idx, cov in enumerate(covariates)]
data_pairs_four= [[covariates[idx], four_four[idx]] for idx, cov in enumerate(covariates)]


xy_kde_one = stats.gaussian_kde(np.transpose(np.array(data_pairs_one)))
xy_kde_two = stats.gaussian_kde(np.transpose(np.array(data_pairs_two)))
xy_kde_three = stats.gaussian_kde(np.transpose(np.array(data_pairs_three)))
xy_kde_four = stats.gaussian_kde(np.transpose(np.array(data_pairs_four)))
```

Let's also visualise what this looks like to make sure it makes sense: 

```python
x_points = np.linspace(0,1,num=100)
y_points = np.linspace(0,3,num=100)

fig = plt.figure(figsize = (10,10))
ax = plt.axes(projection='3d')

X, Y = np.meshgrid(x_points, y_points)
zs = np.array(xy_kde_four.pdf([np.ravel(X), np.ravel(Y)]))
Z = zs.reshape(X.shape)

ax.plot_surface(X,Y,Z)

plt.show()
```

Now we have two density estimates: `x_kde` for the marginal density of the covariates, and `xy_kde_n` for the joint density of bids and covariates. The conditional density is then given by dividing the joint density by the marginal density:

```python
conditional_kde_one = lambda x,y: xy_kde_one.pdf([x,y])/x_kde(x)
conditional_kde_two = lambda x,y: xy_kde_two.pdf([x,y])/x_kde(x)
conditional_kde_three = lambda x,y: xy_kde_three.pdf([x,y])/x_kde(x)
conditional_kde_four = lambda x,y: xy_kde_four.pdf([x,y])/x_kde(x)
```

For our purposes, we need to find the distribution of bids conditional on a realised covariate `x`. This distribution is given by the following function:

```python
def y_distribution_on_x(conditional, x):
    return lambda y: conditional(x,y)

_distribution_one = y_distribution_on_x(conditional_kde_one, 1)
_distribution_two = y_distribution_on_x(conditional_kde_two, 1)
_distribution_three = y_distribution_on_x(conditional_kde_three, 1)
_distribution_four = y_distribution_on_x(conditional_kde_four, 1)
```

```python
values = np.linspace(0,4,num=100)
plt.plot(values, [_distribution_one(x) for x in values], label="p1")
plt.plot(values, [_distribution_two(x) for x in values], label="p2")
plt.plot(values, [_distribution_three(x) for x in values], label="p3")
plt.plot(values, [_distribution_four(x) for x in values], label="p4")
plt.legend()
```

Now we have kernel estimated marginal densities of the distributions of each of the order statistics where `x` is fixed. As with before, to map this to the distribution of the parent distribution, we must first estimate the CDF of these order statistics and map them to the CDF of the parent distribution. Firstly, we reweight the densities to ensure they still integrate to unity, as some precision has been lost during numerical calculations:

```python
weight = integrate.quad(_distribution_one, a=-np.inf, b=np.inf)[0]

distribution_one = lambda y: _distribution_one(y)/weight
distribution_two = lambda y: _distribution_two(y)/weight
distribution_three = lambda y: _distribution_three(y)/weight
distribution_four = lambda y: _distribution_four(y)/weight
```

The CDF of each of the distributions is given by integrating the marginal densities:

```python
KDECDF_one = lambda x: integrate.quad(distribution_one, a=-np.inf, b=x)[0]
KDECDF_two = lambda x: integrate.quad(distribution_two, a=-np.inf, b=x)[0]
KDECDF_three = lambda x: integrate.quad(distribution_three, a=-np.inf, b=x)[0]
KDECDF_four = lambda x: integrate.quad(distribution_four, a=-np.inf, b=x)[0]
```

Now we map these CDFs to the CDF of the parent distribution:

```python
CDF_from_one = get_cdf(KDECDF_one, 1, 4)
CDF_from_two = get_cdf(KDECDF_two, 2, 4)
CDF_from_three = get_cdf(KDECDF_three, 3, 4)
CDF_from_four = get_cdf(KDECDF_four, 4, 4)
```

```python
values = np.linspace(0,4,num=100)

values_from_one = [CDF_from_one(x) for x in values]
values_from_two = [CDF_from_two(x) for x in values]
values_from_three = [CDF_from_three(x) for x in values]
values_from_four = [CDF_from_four(x) for x in values]

plt.plot(values_from_one, label="1;4")
plt.plot(values_from_two, label="2;4")
plt.plot(values_from_three, label="3;4")
plt.plot(values_from_four, label="4;4")

plt.plot(stats.norm.cdf(values, loc=2), label="Actual CDF")

plt.legend()
```

This allows us to then compute the lower and upper bounds, in the same fashion as with the previous working example:

```python
values = np.linspace(0,4,num=100)

def smooth_upper_bound(x, p):
    all_values = [CDF_from_one(x), CDF_from_two(x), CDF_from_three(x), CDF_from_four(x)]
    total = sum([np.exp(g*p) for g in all_values])
    result = sum([g*np.exp(g*p) for g in all_values])/total
    return result

upper_bound_values = [smooth_upper_bound(x, 0) for x in values]
lower_bound_values = [CDF_from_four(x) for x in values]

plt.plot(values, upper_bound_values, label="upper")
plt.plot(values, lower_bound_values, label="lower")

plt.plot(values, stats.norm.cdf(values, loc=2), label="Actual CDF")

plt.legend()
```

As with before, we now discretise the CDF to speed on the calculation time of means:

```python
values = np.linspace(0,4,num=100)

upper_discrete_cdf = [smooth_upper_bound(x, 0) for x in values]
lower_discrete_cdf = [CDF_from_four(x) for x in values]

upper_rv = FastMean(upper_discrete_cdf, values)
lower_rv = FastMean(lower_discrete_cdf, values)

print(f"upper estimate: {lower_rv.mean()}")
print(f"lower estimate: {upper_rv.mean()}")
```

The results are somewhat close to the actual value, 2. It remains for us to condense the code above to a two functions, one that calculates the joint CDF and another that outputs the estimated upper and lower means given a fixed covariate.
